# 7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :
i) Headline
ii) Time
iii) News Link

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page1=requests.get("https://www.cnbc.com/world/?region=world")
```

```python
page1
```

```python
soup1=BeautifulSoup(page1.content,"html.parser")
```

```python
soup1
```

```python
headline=[]
```

```python
for i in soup1.find_all('h2',class_="FeaturedCard-packagedCardTitle"):
    headline.append(i.text)
```

```python
headline
```

```python
for i in soup1.find_all('div',class_="SecondaryCard-headline"):
    headline.append(i.text)
```

```python
headline
```

```python
time=[]
```

```python
for i in soup1.find_all('div',class_="ArticleHeader-time"):
    time.append(i.text)
```

```python
time
```
