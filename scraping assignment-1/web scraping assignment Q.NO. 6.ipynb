# 6) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
**a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.**


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page1=requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
```

```python
page1
```

```python
soup1=BeautifulSoup(page1.content,"html.parser")
```

```python
soup1
```

```python
team1=[]
```

```python
for i in soup1.find_all('span',class_="u-hide-phablet"):
    team1.append(i.text)
```

```python
team1
```

```python
matches1=[]
```

```python
for i in soup1.find_all('td',class_="rankings-block__banner--matches"):
    matches1.append(i.text)
```

```python
matches1
```

```python
for i in soup1.find_all('td',class_="table-body__cell u-center-text"):
    matches1.append(i.text)
```

```python
matches1
```

```python
matches2=matches1[:1]
matches2
```

```python
matches3=matches1[1::2]
matches3
```

```python
matches4=(matches2+matches3)
matches4
```

```python
points1=[]
```

```python
for i in soup1.find_all('td',class_="rankings-block__banner--points"):
    points1.append(i.text)
```

```python
points1
```

```python
for i in soup1.find_all('td',class_="table-body__cell u-center-text"):
    points1.append(i.text)
```

```python
points1
```

```python
points2=points1[::2]
points2
```

```python
rating1=[]
```

```python
for i in soup1.find_all('td',class_="rankings-block__banner--rating u-text-right"):
    rating1.append(i.text)
```

```python
rating1
```

```python
for i in soup1.find_all('td',class_="table-body__cell u-text-right rating"):
    rating1.append(i.text)
```

```python
rating1
```

```python
data7=list(zip(team1,matches4,points2,rating1))
df7=pd.DataFrame(data7,columns=["Team","Match","Points","Rating"])
df7
```
