# 4) Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) 
from https://presidentofindia.nic.in/former-presidents.htm

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page1=requests.get("https://presidentofindia.nic.in/former-presidents.htm")
```

```python
page1
```

```python
soup=BeautifulSoup(page1.content,"html.parser")
```

```python
soup
```

```python
name=[]
```

```python
for i in soup.find_all('div',class_="presidentListing"):
    name.append(i.text)
```

```python
name
```

```python
data=list(name)
df20=pd.DataFrame(data,columns=["Name"])
df20
```
